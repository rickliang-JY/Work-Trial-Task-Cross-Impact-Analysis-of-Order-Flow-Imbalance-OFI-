```python
import pandas as pd 
import numpy as np 
import os
```

```python
##From the reference we only consider the time from 10:00 to 15:30

# 股票名称列表
stock_names = ["AAPL20250102","MSFT20250102","NVDA20250102","AMGN20250102","GILD20250102","TSLA20250102","PEP20250102","JPM20250102", "V20250102", "XOM20250102"]

for stock_name in stock_names:
    # 构造文件路径
    input_file = f'../data/{stock_name}_original.csv'
    output_file = f'../data/{stock_name}_after.csv'
    
    # 读取数据
    data = pd.read_csv(input_file, index_col=1)
    data.index = pd.to_datetime(data.index)
    
    data = data.between_time('10:00', '15:30')
    # 删除指定列
    data = data.drop(columns=['ts_recv', 'rtype', 'publisher_id', 'instrument_id', 'flags'])
    
    # 删除包含数字的列
    data = data.drop(columns=[col for col in data.columns if any(char.isdigit() for char in col) and (int(''.join(filter(str.isdigit, col))) > 5 or int(''.join(filter(str.isdigit, col))) == 0)])
    
    # inorder to make sure that each column has the same length, we fill the missing value with 0
    data.fillna(0, inplace=True)
    #data.dropna(inplace=True)
    
    # 保存处理后的数据
    data.to_csv(output_file)
    
    # 打印前10行
    print(f"Processed {stock_name}:")
    print(data.head(10))
```

```python
data = pd.read_csv('../data/AAPL20250102_original.csv', index_col = 1)
data = data.drop(columns = ['ts_recv','rtype', 'publisher_id', 'instrument_id', 'flags', 'ts_in_delta', 'sequence'])
data = data.drop(columns=[col for col in data.columns if any(char.isdigit() for char in col) and (int(''.join(filter(str.isdigit, col))) > 5 or int(''.join(filter(str.isdigit, col))) == 0)])
data.dropna(inplace=True)
data.to_csv('../data/AAPL20250102_after.csv')
data.head(10)
```
